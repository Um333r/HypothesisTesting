Import Necessary libraries


````python
import pandas as pd
from scipy import stats
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
````

3.

Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.





````python
data_3 = pd.read_csv('/content/BuyerRatio.csv')
data_3
````

````python
data_3.dtypes
````

````python
alpha = 0.05
Males = [50,142,131,70]
Females = [435,1523,1356,750]
Customers = [Males,Females]
print('Customers:',Males,Females)
````

````python
chi2_test_1 = stats.chi2_contingency(Customers)
chi2_test_1
````

````python
print('chi2_value = %f,Pvalue = %f' % (chi2_test_1[0],chi2_test_1[1]))
print('Interpret by Pvalue')
if chi2_test_1[1] < 0.05:
    print('We reject null hypothesis')
else:
    print('we accept null hypothesis')
````

Inference : Pvalue (0.6603) < 0.05 So we accept null hypothesis

Any error in order form renders it defective and has to be reworked before processing. The manager wants to check whether the defective % varies by centre. Please analyze the data at 5% significance level and help the manager draw appropriate inferences

````python
data_4 = pd.read_csv('/content/Costomer+OrderForm.csv')
data_4
````

````python
data_4.head()
````

````python
data_4.describe()
````

````python
data_4.dtypes
````

````python
Phillippines_count = data_4['Phillippines'].value_counts()
print(Phillippines_count)
Indonesia_count = data_4['Indonesia'].value_counts()
print(Indonesia_count)
Malta_count = data_4['Malta'].value_counts()
print(Malta_count)
India_count = data_4['India'].value_counts()
print(India_count)
````

````python
chi2_test_2 = stats.chi2_contingency([[271,267,269,280],[29,33,31,20]])
chi2_test_2
````

````python
print('chi2 = %f,Pvalue = %f' % (chi2_test_2[0],chi2_test_2[1]))
print('Interpret by Pvalue')
if chi2_test_2[1] > 0.05:
    print('We accept null hypothesis')
else:
    print('we reject null hypothesis')
````

Inference : Pvalue (0.2771) > 0.05 So we accept null hypothesis
